# VQVAE Loss Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# for i in ['training_8all.log', 'training_8allKR.log', 'training_12all.log']:
path = '/N/slate/aajais/checkpoints-vqvae-monai-scaled-128/B12-KR-AUG-all-T-256x512|3k|-/training.log'
for i in [path]:
    df = pd.read_csv(i, skiprows=1, header=None)
    # Assign column names
    df.columns = [
        "epoch", "loss", "quantize_loss", "reconst_loss", "perplexity",
        "val_loss", "val_quantize_loss", "val_reconst_loss"
    ]

    # Create subplots
    fig, axs = plt.subplots(4, 1, figsize=(10, 15))

    # Plot loss and validation loss
    axs[0].plot(df['epoch'], df['loss'], label='Training Loss')
    axs[0].plot(df['epoch'], df['val_loss'], label='Validation Loss')
    axs[0].set_xlabel('Epoch')
    axs[0].set_ylabel('Loss')
    axs[0].set_title('Training and Validation Loss ' + i)
    axs[0].legend()
    axs[0].grid(True)

    # Plot quantize loss and validation quantize loss
    axs[1].plot(df['epoch'][10:], df['quantize_loss'][10:], label='Training Quantize Loss')
    axs[1].plot(df['epoch'][10:], df['val_quantize_loss'][10:], label='Validation Quantize Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].set_ylabel('Quantize Loss')
    axs[1].set_title('Training and Validation Quantize Loss ' + i)
    axs[1].legend()
    axs[1].grid(True)

    # Plot reconstruction loss and validation reconstruction loss
    axs[2].plot(df['epoch'], df['reconst_loss'], label='Training Reconstruction Loss')
    axs[2].plot(df['epoch'], df['val_reconst_loss'], label='Validation Reconstruction Loss')
    axs[2].set_xlabel('Epoch')
    axs[2].set_ylabel('Reconstruction Loss')
    axs[2].set_title('Training and Validation Reconstruction Loss '+ i)
    axs[2].legend()
    axs[2].grid(True)
    
    # Plot perplexity
    axs[3].plot(df['epoch'], df['perplexity'], label='Training Perplexity')
    axs[3].set_xlabel('Epoch')
    axs[3].set_ylabel('Perplexity')
    axs[3].set_title('Training and Validation Reconstruction Loss '+ i)
    axs[3].legend()
    axs[3].grid(True)

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()
```
